```python
import matplotlib.pyplot as plt
years = ["1st sem", "2nd sem", "3rd sem", "4th sem"]
Priti = [80, 89, 93, 88]
abc = [70, 80, 81, 86]
xyz = [71, 80, 83, 87]


plt.plot(years, Priti, label="Priti", color="purple")
plt.plot(years, abc, label="abc", color="pink")
plt.plot(years, xyz, label="xyz", color="blue")


plt.xlabel("Semester")
plt.ylabel("Percentage Scored")
plt.title("performance comparison")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

semesters = ['Sem 1', 'Sem 2', 'Sem 3', 'Sem 4']
marks = [80.59, 89.40, 93.06, 88.40]

styles = [
    'default', 'classic', 'seaborn-v0_8', 'ggplot', 'bmh',
    'fivethirtyeight', 'Solarize_Light2', 'fast', 'dark_background', 'grayscale'
]

for style in styles:
    plt.style.use(style)
    plt.figure(figsize=(8, 5))
    plt.bar(semesters, marks, color='purple', edgecolor='black')
    plt.title(f'Priti\'s Semester Marks - Style: {style}')
    plt.ylabel('Marks')
    plt.xlabel('Semesters')
    plt.ylim(0, 100)
    plt.grid(True, linestyle='--')
    plt.show()
```

```python
years = ['1990', '1995', '2000', '2005', '2010', '2015', '2020']
kohli = [0, 0, 500, 800, 1100, 1300, 1500]
Sachin = [0, 300, 800, 1200, 1500, 1700, 1600]
Dhoni = [59, 0, 400, 800, 1200, 1100, 1580]
plt.xlabel('Years')
plt.ylabel('Runs scored')
plt.plot(years, kohli, color='orange', linestyle='-.', linewidth=3, label="Kohli")
plt.plot(years, Sachin, color='green', linestyle='-.', label="Sachin")
plt.plot(years, Dhoni,'ro--', label="Dhoni")
plt.grid(True)
plt.tight_layout()
plt.legend()
```
